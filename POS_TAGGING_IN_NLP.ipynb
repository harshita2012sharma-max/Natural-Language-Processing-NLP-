##**POS tagging**

```python
# pos tagging means parts of speech tagging.
# It is NLP tr=echinique where each word in a
# sentence is assigned its grammatical role such as:

# Noun
# Verb
# Adjective
# Pronoun
# Preprosition
# Conjuction
```

```python
# WHY IS POS TAGGING IMPORTANT?

# POS Taggingis used in:
#  Text Classification
#ChatBocats
# Machine Translation
# Nmaed Entity Recognisation
# Setiment Analysis
# $ Speech Recognisation
# Grammer checking
```

## **Commmon POS Tagging**

```python
# |POS Tagging    |  Meaning       |
# |---------------|--------------- |
# |NN             | Noun           |
# |NNS            | Plural Noun    |
# |NNP            | Proper Noun    |
# |VB             | Verb           |
# |VBG            | Verb Gerund    |
# |VBD            | Verb Past Tense|
# |JJ             | Adjective      |
# |RB             | Adverb         |
# |PRP            | Pronoun        |
# |IN             | Adjective      |
```

```python
#  Working Process of POS Tagging ::::

# 1. Sentence is divided into tokens.
# 2. NLP model checks grammar/context
# 3. Appropriate POS tag is assigned to each word.
```

```python
import pandas as pd

df = pd.read_csv(
    "/content/IMDB Dataset - IMDB Dataset.csv",
    engine='python'
)

df.head()
```

```python
# Clean
```

```python
df['review']=df['review'].str.lower()
```

```python
# Remove html Tags
```

```python
import re
```

```python
def remove_html_tags(text):
  pattern=re.compile('<.*?>')
  return pattern.sub(r'',text)
df['review']=df['review'].apply(remove_html_tags)
df.head()
```

```python
# remove URL
```

```python
def remove_url(text):
  pattern=re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)

df['review']=df['review'].apply(remove_url)
```

```python
# Remove punctuation
```

```python
import string
print(string.punctuation)

exclude=string.punctuation
```

```python
def remove_punct(text):
  for char in exclude:
    text=text.replace(char,"")
  return text
df['review']=df['review'].apply(remove_punct)
df.head()
```

```python
# Spelling correction
```

```python
from textblob import TextBlob

text = df['review'][0]

blob = TextBlob(text)

print(blob.correct())
```

```python
import re
import string

text = df['review'][0]

# lowercase
text = text.lower()

# remove html tags
text = re.sub(r'<.*?>', '', text)

# remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# remove numbers
text = re.sub(r'\d+', '', text)

# remove extra spaces
text = re.sub(r'\s+', ' ', text).strip()

print(text)
```

```python
text
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")

doc=nlp(text)
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
from spacy import displacy

import spacy
nlp=spacy.load("en_core_web_sm")

doc=nlp(text)
displacy.render(doc, style="ent", jupyter=True)
```
