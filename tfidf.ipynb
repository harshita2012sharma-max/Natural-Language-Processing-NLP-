```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'text':['People watch cricket','player was player','people write comment'
,'cricket play comment'],'output':[1,1,0,0]})
```

```python
#people watch cricket player was write comment play
#[11100000]
#[00021000]
#[10000110]
#[00100011]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
#vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
```

```python
cv.transform(['cricket watch and write comment of cricket comment']).toarray()
```

```python
#TF(term Frequency)
# 2. IDF(Inverse Documnet Frequency)
```

```python
#tf(t,d)=(Numer of occuracnces of term t in document d)/(Total number of term in the document d)

# example -->
# d1=people watch cricket.

# tf(people,d1)=1/3

# d2=match was match
# tf(match,d2)=2/3
```

```python
#IDF(t) =Loge(Total number of document in the corpus)/(Number of documents with term t in them)

# here corpus means number of sentences

# Examples ==>
# d1 = people watch cricket
# d2=cricket was cricket
#d3=people write comment
# d4 =cricket write comment

#idf(cricket)=Log(4/4)
#idf(people)=Log(4/2)
```
