Stemming

```python
#1. Inflection =>  "IN grammer ,inflection is the madification of a word to express
# different gramatical categories such ad tense ,case,gevder,voice
# .aspect,person,number,mood"

# 2. Stemming =>"" Stemming is the process of redicing the 
# inflection in the words to their root  form such as mapping a group of words to the 
# same stem even if the stem itself is not valid word in language"
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample="walk walks walked walking"
stem_words(sample)
```

```python
#stemming mapping a group of words to the same stem if the stem itself 
# is not valid word in language.So.this is the big issue of stemming
```

```python
text='probably my alltime favourite movie a story of sleffessness scarifice and dedication.'
print(text)
stem_words(text)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
import string

wnl=WordNetLemmatizer()

sentence="He was running and eating at the same time ,he had a bad habit of swimming afetr the long hour of playing in Sun."
punctuation=string.punctuation
sentence_words=nltk.word_tokenize(sentence)

for word in sentence_words:
    if word in punctuation:
        sentence_words.remove(word)

sentence_words

print('{0:20}{1:20}'.format("Word","Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word,wnl.lemmatize(word,pos='v')))
```
